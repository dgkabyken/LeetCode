```python
# Solution 1
class Solution:
    def getSum(self, a: int, b: int) -> int:
        def summ(a, b):
            tmp = 0
            while (a != 0) and (b != 0):
                tmp = (a&b)<<1
                b = (a^b)
                a = tmp
            return a^b
        def subs(a, b):
            tmp = 0
            while (a != 0) and (b != 0):
                tmp = (~a)&b
                a = a^b
                b = tmp<<1
            return a^b
        def func(a, b):
            if (a >= 0) & (b >= 0):
                return summ(a, b)
            if (a <= 0) & (b <= 0):
                return summ(a, b)
            if (a >= 0) & (b <= 0):
                if abs(a) < abs(b):
                    return summ(a, b)
                else:
                    return subs(a, ~b+1)
            if (a <= 0) & (b >= 0):
                if abs(a) > abs(b):
                    return summ(a, b)
                else:
                    return subs(b, ~a+1)
        return func(a, b)
```
